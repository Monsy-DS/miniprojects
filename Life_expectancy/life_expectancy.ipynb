```python
import plotly.plotly as py
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import pandas as pd
```

```python
init_notebook_mode(connected=True)
```

```python
df = pd.read_csv('Life Expectancy Data.csv',sep=',', encoding='latin-1' )
```

```python
df.head()
```

```python
df.columns
```

```python
df2 = df[['Country', 'Life expectancy ', 'GDP']]
```

```python
df3 = df2.groupby(by='Country').mean()
```

```python
df3.head()
```

```python
GDP = pd.read_csv('2014_World_GDP')

GDP.columns=['Country', 'GDP', 'Code']
GDP = GDP.drop(['GDP'], axis=1)
GDP = GDP.set_index('Country')
GDP.head()
```

```python
df4 = pd.merge(df3, GDP, left_index=True, right_index=True)
```

```python
df4.head()
```

```python
df4.columns = ['Life_expectancy','GDP','Code']
```

```python
df4 = df4.reset_index()
```

```python
data = dict(type = 'choropleth',
           locations=df4['Code'],
           z=df4['Life_expectancy'],
            text = df4['Country'],
           colorbar={'title':'Life_expectancy'})
layout = dict(title = 'Life expectancy',
             geo = dict(showframe = False,
                        projection = {'type':"natural earth"},
                        showocean=True, oceancolor='rgb(100,255,255)'
             ))
choromap = go.Figure(data = [data], layout=layout)
iplot(choromap)
```

```python
data = dict(type = 'choropleth',
           locations=df4['Code'],
           z=df4['GDP'],
            text = df4['Country'],
           colorbar={'title':'GDP'})
layout = dict(title = 'GDP',
             geo = dict(showframe = False,
                        projection = {'type':"natural earth"},
                        showocean=True, oceancolor='rgb(100,255,255)'
             ))
choromap = go.Figure(data = [data], layout=layout)
iplot(choromap)
```
